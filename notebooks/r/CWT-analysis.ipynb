```r
pair = preprocessing('E:/summary/sept_casual.csv', 'E:/summary/sept_total.csv')
sept_cas = ts(pair$series[25:696], frequency=24)
sept_tot = ts(pair$total[25:696], frequency=24)
```

```r
june_cas = ts(preprocessing('E:/summary/june_casual.csv', 'E:/summary/june_total.csv')$series[49:720], frequency=24)
june_tot = ts(preprocessing('E:/summary/june_casual.csv', 'E:/summary/june_total.csv')$total[49:720], frequency=24)
```

```r
sept_perc = sept_cas/sept_tot
```

```r
june_perc = june_cas/june_tot
```

```r
sept_cwt = wavCWT(sept_perc, wavelet='gaussian2', n.scale = 800, scale.range =c(0.05, 7))
cwt.s= as.matrix(sept_cwt)
```

```r
june_cwt = wavCWT(june_perc, wavelet='gaussian2', n.scale = 800, scale.range =c(0.05, 7))
cwt.j= as.matrix(june_cwt)
```
